```python
import pandas as pd

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)
df.head(8)
```

```python
import matplotlib.pyplot as plt


df['Salário'] = pd.to_numeric(df['Salário'])
df['Limite_Credito'] = pd.to_numeric(df['Limite_Credito'])

df_sorted = df.sort_values(by=['Salário', 'Limite_Credito'], ascending=[False, False])

plt.figure(figsize=(10,7))
plt.scatter(df_sorted['Salário'], df_sorted['Limite_Credito'], color='blue')

for i, row in df_sorted.iterrows():
    plt.text(row['Salário'] + 100, row['Limite_Credito'], f"RS{row['Salário']} / R${row['Limite_Credito']}", fontsize=9)

plt.title('Relação entre Salário e Limite de Crédito (Ordenado do Maior para o Menor)')
plt.xlabel('Salário (R$)')
plt.ylabel('Limite de Crédito (R$)')
plt.grid(True)
plt.show()

#Nota-se que existe uma correlação positiva entre renda e limites de crédito;
#Caracteriza que consumidores de maior renda possuem maiores limite de crédito;
#Porém muitos há muitas variáveis que potencializam e represam limites de crédito;
#Dentre elas: Histórico de Inadimplência, Posse de imovel, solicitação de Aumento.
```

```python
import seaborn as sns

plt.figure(figsize=(10,7))
sns.barplot(data=df, x='Faixa_Etaria', y='Limite_Credito', estimator='mean', ci=None)
plt.title('Média do Limite de Crédito por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Média do Limite de Crédito')
plt.show()

#Um fator impulsionador de crédito é idade e boa relações com contas;
#Maior estabilidade financeira e histórico de crédito positivo com lojas e bancos;
#Quanto mais giro de Capital em crédito maior se torna seu limite de crédito;
#Logo pessoas que gastam e não buscam ampliar podem sim se manter na mesma faixa etária.
#É inteligente usar o crédito com responsabilidade e aumentar gradativamente ao longo da vida.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    'Nome': ['Alice', 'João', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico', 'Engenheiro', 'Estudante', 'Estudante'],
    'Salário': [4500, 8000, 5000, 10000, 12000, 15000, 1200, 1500],
    'Limite_Crédito': [2500, 4000, 4000, 1000, 10000, 2000, 500, 250],
    'Histórico_Inadimplência': [0, 0, 0, 1, 0, 1, 0, 1],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento', 'Solteiro', 'Solteiro', 'Solteiro'],
    'Imovel_Proprio': [0, 0, 0, 1, 1, 1, 0, 0]
}

df = pd.DataFrame(data)
fig, ax = plt.subplots(figsize=(10, 6))

for index, row in df.iterrows():
    color = 'red' if row['Histórico_Inadimplência'] == 1 else 'blue'
    ax.bar(row['Nome'], row['Limite_Crédito'], color=color, label='Crédito' if index == 0 else "")
    ax.bar(row['Nome'], row['Salário'], bottom=row['Limite_Crédito'], color='gray', label='Salário' if index == 0 else "")

ax.set_ylabel("Valores")
ax.set_xlabel("Nome")
ax.set_title("Gráfico de Salário e Limite de Crédito por Pessoa")
ax.legend(["Crédito (Azul/Inadimplentes Vermelho)", "Salário"], loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# A passagem por um histórico de inadimplência e nome sujo pode represar seu limite de crédito;
# A concessão de crédito é feita por instituições financeiras e são micro empréstimos.
# Uma vez que o banco perde a confiança sua concessão de crédito se torna mais difícil.
# Como visto no gráfico, mesmo salários maiores com histórico possuem limites menores que salários menores.
```
